```python
import numpy as np 
import pandas as pd 
import matplotlib as plt
import seaborn as sb
```

```python
sb.set_palette("husl",3)
sb.set_style("whitegrid")
```

```python
current_sample = pd.read_csv("../server/data/sample_wrong_0.txt", sep=" ", names=["elapsedTime","fsr_1","fsr_2","accelx_1","gyroy_1","gyroz_1","gyrox_2","gyroy_2","gyroz_2","gyrox_3","gyroy_3","gyroz_3","gyrox_4","gyroy_4","gyroz_4","accelx_1","accely_1","accelz_1","accelx_2","accely_2","accelz_2","accelx_3","accely_3","accelz_3","accelx_4","accely_4","accelz_4"])
print ('Number of samples: '+ str(len(current_sample)))
print (current_sample)
current_sample_pairplot = sb.pairplot(current_sample)
```

```python
print(current_sample.columns)
current_sample.describe()
```

Plot All Gyro Data

```python

gyro_current_sample_1 = current_sample[['elapsedTime','gyrox_1', 'gyroy_1', 'gyroz_1']]
gyro_melted_current_sample_1 = gyro_current_sample_1.melt('elapsedTime',var_name='Gyro''s',value_name='Rad/Sec')

gyro_current_sample_2 = current_sample[['elapsedTime','gyrox_2', 'gyroy_2', 'gyroz_2']]
gyro_melted_current_sample_2 = gyro_current_sample_2.melt('elapsedTime',var_name='Gyro''s',value_name='Rad/Sec')

gyro_current_sample_3 = current_sample[['elapsedTime','gyrox_3', 'gyroy_3', 'gyroz_3']]
gyro_melted_current_sample_3 = gyro_current_sample_3.melt('elapsedTime',var_name='Gyro''s',value_name='Rad/Sec')

gyro_current_sample_4= current_sample[['elapsedTime','gyrox_4', 'gyroy_4', 'gyroz_4']]
gyro_melted_current_sample_4 = gyro_current_sample_4.melt('elapsedTime',var_name='Gyro''s',value_name='Rad/Sec')

current_sample_gyro_1 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Gyro''s', data=gyro_melted_current_sample_1, kind='line')
current_sample_gyro_1.set(ylim=(-5.5,5.5))

current_sample_gyro_2 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Gyro''s', data=gyro_melted_current_sample_2, kind='line')
current_sample_gyro_2.set(ylim=(-5.5,5.5))

current_sample_gyro_3 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Gyro''s', data=gyro_melted_current_sample_3, kind='line')
current_sample_gyro_3.set(ylim=(-5.5,5.5))

current_sample_gyro_4 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Gyro''s', data=gyro_melted_current_sample_4, kind='line')
current_sample_gyro_4.set(ylim=(-5.5,5.5))
```

Plot All Accel Data

```python
accel_current_sample_1 = current_sample[['elapsedTime','accelx_1', 'accely_1', 'accelz_1']]
accel_melted_current_sample_1 = accel_current_sample_1.melt('elapsedTime',var_name='Accel''s',value_name='Rad/Sec')

accel_current_sample_2 = current_sample[['elapsedTime','accelx_2', 'accely_2', 'accelz_2']]
accel_melted_current_sample_2 = accel_current_sample_2.melt('elapsedTime',var_name='Accel''s',value_name='Rad/Sec')

accel_current_sample_3 = current_sample[['elapsedTime','accelx_3', 'accely_3', 'accelz_3']]
accel_melted_current_sample_3 = accel_current_sample_3.melt('elapsedTime',var_name='Accel''s',value_name='Rad/Sec')

accel_current_sample_4= current_sample[['elapsedTime','accelx_4', 'accely_4', 'accelz_4']]
accel_melted_current_sample_4 = accel_current_sample_4.melt('elapsedTime',var_name='Accel''s',value_name='Rad/Sec')

current_sample_accel_1 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Accel''s', data=accel_melted_current_sample_1, kind='line')
current_sample_accel_1.set(ylim=(-20,20))

current_sample_accel_2 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Accel''s', data=accel_melted_current_sample_2, kind='line')
current_sample_accel_2.set(ylim=(-20,20))

current_sample_accel_3 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Accel''s', data=accel_melted_current_sample_3, kind='line')
current_sample_accel_3.set(ylim=(-20,20))

current_sample_accel_4 = sb.relplot(x="elapsedTime", y="Rad/Sec", hue='Accel''s', data=accel_melted_current_sample_4, kind='line')
current_sample_accel_4.set(ylim=(-20,20))
```

Plot All FSR Data

```python
fsr_current_sample = current_sample[['elapsedTime','fsr_1', 'fsr_2']]
fsr_melted_current_sample = fsr_current_sample.melt('elapsedTime',var_name='FSR''s',value_name='Data/Sec')

current_sample_fsr = sb.relplot(x="elapsedTime", y="Data/Sec", hue='FSR''s', data=fsr_melted_current_sample, kind='line')
current_sample_fsr.set(ylim=(0,200))
```
